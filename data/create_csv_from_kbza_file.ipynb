```python
import pandas as pd

filepath = 'resultados_exp.txt'
diff = 0
nodes = 0
df_list = []
with open(filepath) as fp:
    for line in fp:
        if 'diff' in line:
            _, str_diff, _, str_nodes = line.rstrip().split(' ')
            difficulty = int(str_diff)
            nodes = int(str_nodes)

        if 'real' in line:
            time = float(line.rstrip().split(' ')[-1][2:-1])
            d = {'difficulty':difficulty, 'nodes':nodes, 'time': time}
            df = pd.DataFrame([d])
            df_list.append(df)
exp_df = pd.concat(df_list)
exp_df
```

```python
exp_df.to_csv('exp')
```
